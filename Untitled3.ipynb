```python
'''
문제 1번
'''from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['class'] = cancer.target
df.head() #우선 데이터를 한번출력한다.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='class',data=df)
plt.show() #데이터 불균형 정도를 한번 확인해보자
```

```python
from sklearn.model_selection import train_test_split as ttspl
x = df.iloc[:,1:-1]
y = df.iloc[:,-1]
x_train,x_test,y_train,y_test =ttspl(x,y,test_size=0.3,random_state=40,stratify=y)#학습,검증데이터 나누기
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
def get_clf_eval(y_test,pred):  #각 평가지표를 반환하는 함수
  acc = accuracy_score(y_test,pred)
  pre = precision_score(y_test,pred)
  re = recall_score(y_test,pred)
  f1 = f1_score(y_test,pred)
  auc = roc_auc_score(y_test,pred)
  return acc,pre,re,f1,auc
def print_clf_eval(y_test,pred):  # 평가지표들을 출력하는 함수
  confusion = confusion_matrix(y_test,pred)
  acc,pre,re,f1,auc =  get_clf_eval(y_test,pred)
  print("오차행렬:")
  print(confusion)
  print("=========")
  print("정확도:{0:.4f},정밀도{1:.4f}".format(acc,pre))
  print("재현:{0:.4f}, F1:{1:.4f}, AUC:{2:.4f}".format(re,f1,auc))


from sklearn.linear_model import LogisticRegression        #각 모델의 평가지표 출력
lr_clf = LogisticRegression(random_state=40,solver='liblinear')
lr_clf.fit(x_train,y_train)
lr_pred = lr_clf.predict(x_test)
print("Logistic regression")
print_clf_eval(y_test,lr_pred)

print("\n\n")

from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(random_state=40,max_depth=3)
dt_clf.fit(x_train,y_train)
dt_pred = dt_clf.predict(x_test)
print("Decision Tree")
print_clf_eval(y_test,dt_pred)

print("\n\n")

from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier(random_state=40,n_jobs=-1,n_estimators=100)
rf_clf.fit(x_train,y_train)
rf_pred = rf_clf.predict(x_test)
print("Random Forest")
print_clf_eval(y_test,dt_pred)

print("\n\n")

from lightgbm import LGBMClassifier
lgbm_clf = LGBMClassifier(n_estimators=1000,num_leaves=64,n_jobs=-1,boost_from_average=False)
lgbm_clf.fit(x_train,y_train)
lgbm_pred = rf_clf.predict(x_test)
print("LightGBM")
print_clf_eval(y_test,dt_pred)
```

```python
'''
문제 2번
'''from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

pred_proba_1 = rf_clf.predict_proba(x_test)[:,1]  #랜덤 포레스트 모델의 ROC 커브
fall_out,recall,thresholds = roc_curve(y_test,pred_proba_1)

plt.figure(figsize = (8,8))
plt.plot(fall_out,recall,label="Random Forest")
plt.title("ROC,AUC : "+ str(roc_auc_score(y_test,pred_proba_1)))
plt.show
```

```python
from sklearn.metrics import precision_recall_curve

precision,recall,threshold = precision_recall_curve(y_test,pred_proba_1)

plt.figure(figsize = (8,8))    #랜텀포레스트 모델의 recall, precision curve
plt.plot(threshold,precision[:len(threshold)],label = 'precison')
plt.plot(threshold,recall[:len(threshold)],label = 'recall')
plt.legend()
plt.grid()
plt.xlabel('threshold')
plt.show
```
